```python
height=[174,150,180,155,160]
weight=[70,53,75,60,65]
plt.subplot(1,2,1)
plt.plot(weight,height,marker='s',mfc='r',mec='b',linestyle='dotted',linewidth=5,color='k')
plt.xlabel('weight')
plt.ylabel('height')
plt.suptitle('height vs weight')
plt.grid()

plt.Subplot(1,2,2)
plt.plot(height,weight,marker='o',mec='y',mfc='g',linestyle='dashed',linewidth=3,color='b')
plt.ylabel('weight')
plt.xlabel('height')
plt.suptitle('weight vs height ')
plt.grid()
```
